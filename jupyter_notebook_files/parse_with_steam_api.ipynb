```python
import json
import requests
import pandas as pd


data = dict()
base_url = "https://api.steampowered.com/ISteamApps/GetAppList/v2/"
r = requests.get(base_url)
data = json.loads(r.content)
data = data["applist"]["apps"]
```

```python
steam_games = pd.DataFrame.from_dict(data)
steam_games.head()
```

```python
import string

def change_name(x):
    x = x.lower()
    
    new_name = str()
    for i in x:
        if i in (string.ascii_letters + string.digits + " " + "РрАаБбВвГгДдЕеЁёЖжЗзИиЙйКкЛлМмНнОоПпСсТтУуФфХхЦцЧчШшЩщЪъЫыЬьЭэЮюЯя"):
            new_name += i

    return " ".join(new_name.split())

df = pd.read_csv("T_games_dataset.csv")
games = pd.DataFrame(dict(name=pd.unique(df.good_name)))
games = games.merge(df.good_name.value_counts().reset_index(), left_on="name", right_on="good_name")
# games.name = games.name.apply(lambda x: x[:-6] if x.endswith("DLC") else x)
games.name = games.name.apply(change_name)

def delete_dlc(x):
    if x.endswith("dlc"):
        return x[:-4]

    return x

# print(games[games.name.str.endswith("dlc")])
games.name = games.name.apply(delete_dlc)
steam_games.name = steam_games.name.apply(change_name)

# games[games.name.str.endswith("dlc")]
```

```python
r = games.merge(steam_games, on="name", how="left")
r[r.appid.isna() & (r["count"] > 100)]
```
