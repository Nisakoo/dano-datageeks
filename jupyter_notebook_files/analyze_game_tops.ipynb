```python
import pandas as pd


game_top = pd.read_csv("steam_top_4171.csv")
df = pd.read_csv("T_games_dataset.csv")
df.head(1)
```

```python
t_games = pd.DataFrame(dict(name=pd.unique(df.good_name)))
t_games = t_games.merge(game_top, on="name", how="left")

print(t_games.shape)
print(t_games[t_games.rating.isna()].shape)

r = df[df.steam_popularity_score > 0]
r = r.drop(r[r.good_name.duplicated()].index)

t_games = t_games.merge(r[["good_name", "steam_popularity_score"]], left_on="name", right_on="good_name")
t_games[~t_games.rating.isna()]
```
