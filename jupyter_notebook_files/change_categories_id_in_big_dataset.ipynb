```python
import pandas as pd


df = pd.read_csv("T_games_dataset_full_categories.csv")
df = df.drop(columns=["Unnamed: 0"])
```

```python
from functools import lru_cache

old_df = pd.read_csv("T_games_dataset.csv")

@lru_cache(None)
def get_category_id(x):
    category_id = list(old_df[old_df.category_name == x].category_id)[0]
    return category_id

categories = pd.DataFrame({"category": pd.unique(old_df.category_name)})
categories["id"] = categories.category.apply(get_category_id)
categories
```

```python
@lru_cache(None)
def get_category_id(x):
    return int(categories[categories.category == x].id)

df.category_id = df.apply(lambda x: get_category_id(x["category_name"]), axis=1)
```

```python
df.to_csv("T_games_dataset_full_categories.csv", index=False)
```
